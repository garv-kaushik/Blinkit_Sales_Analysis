```python
#Data cleaning

blinkit_df['Item Fat Content'] = blinkit_df['Item Fat Content'].replace({'LF' : 'Low Fat',
                                                                      'low fat' : 'Low Fat',
                                                                      'reg' : 'Regular'})
```

# **Business Requirments**

## **KPI's Requirments**

```python
#Total sales
total_sales = blinkit_df['Sales'].sum()

#Average sales
avg_sales = blinkit_df['Sales'].mean()

#No. of item sold
no_of_item_sold = blinkit_df['Sales'].count()

#Average Ratings
avg_rating = blinkit_df['Rating'].mean()
```

```python
#Display Output

print(f"Total Sales : ${total_sales:,.0f}")
print(f"Average Sales : ${avg_sales:,.0f}")
print(f"NO. Of Item Sold : {no_of_item_sold:,.0f}")
print(f"Average Rating : {avg_rating:,.0f}")
```

# **Charts Requirments**

```python
#Total Sales by Fat Content

sales_by_fat = blinkit_df.groupby('Item Fat Content')['Sales'].sum()
plt.pie(sales_by_fat, labels=sales_by_fat.index, autopct='%1.1f%%', startangle=90, shadow=True )
plt.title('Pie Chart For Sales By Fat Content')
plt.axis('equal')
plt.show()
```

**The pie chart shows that products with **low fat** account for a majority of Blinkit's sales at 64.6%. Regular fat products make up the remaining 35.4%, indicating a strong customer preference for healthier, low-fat options.**

```python
#Total Sales By Item
# blinkit_df['Item Type'].unique()

sales_by_item = blinkit_df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
bars = plt.bar(sales_by_item.index, sales_by_item.values);  # <--- semicolon here

plt.xticks(rotation=-90)
plt.xlabel('Item Type')
plt.ylabel('Sales')
plt.title('Bar Chart For Sales By Item Type')

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f'{bar.get_height():.0f}',
        ha='center', va='bottom', fontsize=8
    )

plt.tight_layout()
plt.show()
```

**Fruits and vegetables, followed closely by snack foods, are the top-selling item types on Blinkit, with both categories having sales above 175,000 units. Seafood and breakfast items are the least popular, indicating lower customer demand for these categories.**

```python
#Total Sales By Outlet By Their Fat Content

grouped = blinkit_df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Low Fat', 'Regular']]

ax = grouped.plot(kind='bar', figsize=(6,5), title='Outlet Teir By Fat Content')
plt.xlabel('Outlet Location Type')
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')
plt.tight_layout()
plt.show();
```

**Low fat items consistently outsell regular fat items across all outlet tiers, with the sales gap increasing from Tier 1 to Tier 3 locations. This suggests a stronger consumer shift toward low fat products, especially in lower-tier markets.**

```python
#Total Sales By Outlet Establishment

sales_by_year = blinkit_df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(10,6))
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-')

# Add values on each marker
for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom', fontsize=9)

plt.title('Outlet Establishment')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.tight_layout()
plt.show()
```

**Outlets established before 2000 show the highest total sales, while those established around 2010 experienced a significant dip in sales performance. Sales have generally stabilized for outlets opened after 2010, maintaining consistent figures around 130,000 units.**

```python
#Sales by Outlet Size

sales_by_outlet_size = blinkit_df.groupby('Outlet Size')['Sales'].sum()
plt.figure(figsize=(4,4))

plt.pie(sales_by_outlet_size, labels=sales_by_outlet_size.index, autopct='%1.1f%%', startangle=90, shadow=True )
plt.title('Outlet Size')
plt.tight_layout()
plt.show()
```

**Medium-sized outlets represent the largest segment, accounting for 42.3% of the total, followed by small outlets at 37%. High-sized outlets form the smallest group at 20.7%, indicating Blinkit's operations are primarily concentrated in medium and small locations.**


```python
#Sales By Outlet Location

sales_by_outlet_locaton = blinkit_df.groupby('Outlet Location Type')['Sales'].sum().sort_values( ascending=False).reset_index()

plt.figure(figsize=(9,3))
sns.barplot(data=sales_by_outlet_locaton, x='Sales', y='Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location')
plt.tight_layout()
plt.show();
```

**Tier 3 outlets achieve the highest total sales among all locations, outperforming both Tier 1 and Tier 2 outlets. Tier 1 locations record the lowest total sales, indicating greater sales activity in less urbanized areas.**
